### 4. Plotting a choropleth

```python
# Create a data frame with just the states and the values for rating we want plotted
data_to_plot = df[['Country','City','AQI Value']]
data_to_plot.head()
```

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [40, -95], zoom_start = 4)

folium.Choropleth(
    geo_data = country_geo, 
    data = data_to_plot,
    columns = ['Country', 'AQI Value'],
    key_on = 'name_en', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'YlGnBu', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "rating").add_to(map)

folium.LayerControl().add_to(map)

map
```

```python
#save the map
map.save('plot_data.html')
```
